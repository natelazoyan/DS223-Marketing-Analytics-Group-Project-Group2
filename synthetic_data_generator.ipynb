## Loading modules and packages

```python
from .data_generator import generate_property
from .data_generator import generate_customer
from .data_generator import generate_agent
from .data_generator import generate_transaction
from .data_generator import generate_date
from .data_generator import generate_sales
import pandas as pd
import random
```

## Declaring Constants

```python
NUMBER_OF_PROPERTIES=5000
NUMBER_OF_CUSTOMERS=3000
NUMBER_OF_AGENTS=2000
NUMBER_OF_TRANSACTIONS=4000
```

## Generating Customer Data

```python
customer_data = [generate_customer(customer_id) for customer_id in range(NUMBER_OF_CUSTOMERS)]

pd.DataFrame(customer_data)
#.to_csv('M2/DB/data_csv/customer.csv', index=False)
```

## Generating Agent Data

```python
agent_data = [generate_agent(agent_id) for agent_id in range(NUMBER_OF_AGENTS)] 
pd.DataFrame(agent_data)
#.to_csv('M2/DB/data_csv/agent.csv', index=False)
```

## Generating Property Data

```python
property_data = [generate_property(property_id) for property_id in range(NUMBER_OF_PROPERTIES)]
pd.DataFrame(property_data)
#.to_csv('M2/DB/data_csv/property.csv', index=False)
```

## Generating Transaction Data

```python
transaction_data = [generate_transaction(transaction_id) for transaction_id in range(NUMBER_OF_TRANSACTIONS)] 
pd.DataFrame(transaction_data)
#.to_csv('M2/DB/data_csv/transactions.csv', index=False)
```

## Generating Dates Data

```python
from datetime import datetime

# Define the start and end dates as strings
start_date = "2000-01-01"
end_date = "2023-12-31"

# Convert the date strings to datetime objects
start_date_obj = datetime.strptime(start_date, "%Y-%m-%d")
end_date_obj = datetime.strptime(end_date, "%Y-%m-%d")

# Calculate the difference between the two dates
number_of_days = (end_date_obj - start_date_obj).days
number_of_days
```

```python
dates_data = [generate_date(date_id) for date_id in range(number_of_days+1)] 
pd.DataFrame(dates_data)
#.to_csv('M2/DB/data_csv/date.csv', index=False)
```

## Generating Sales Data

```python
sales_data = []


for transaction_id in range(1, NUMBER_OF_TRANSACTIONS + 1):
    property_id = random.randint(1, len(property_data))
    customer_id = random.randint(1, len(customer_data))
    agent_id = random.randint(1, len(agent_data))
    date_id = random.randint(1, len(dates_data))
    
    sale = generate_sales(transaction_id, property_id, customer_id, agent_id, date_id)
    sales_data.append(sale)
```

```python
pd.DataFrame(sales_data)
#.to_csv('M2/DB/data_csv/sales.csv', index=False)
```
